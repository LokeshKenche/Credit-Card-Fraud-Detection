```python
non_fraud=data[data['Class']==0]
fraud=data[data['Class']==1]
```

```python
fraud_percentage=len(fraud)/len(data)*100
print(fraud_percentage)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
data['Normalized_amount']=scaler.fit_transform(data['Amount'].values.reshape(-1,1))
data=data.drop(['Amount','Time'],axis=1)
```

```python
data.head()
```

```python
#for all independent varaiables
x=data.drop(['Class'],axis=1)
#for all dependent
y=data['Class']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#creating model
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
x_train_imputed = imputer.fit_transform(x_train)
x_test_imputed = imputer.transform(x_test)
# Handling missing values in the target variable (y_train)
y_train = y_train.dropna() # Droping rows with missing target values
y_train = y_train.reset_index(drop=True)
x_train_imputed = x_train_imputed[y_train.index]

rf = RandomForestClassifier(n_estimators=100)
rf.fit(x_train_imputed, y_train)
```

```python
prediction_rf=rf.predict(x_test_imputed)
Score=rf.score(x_test_imputed,y_test)*100
print(Score)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction_rf))
```
